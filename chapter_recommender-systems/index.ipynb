
# 推薦システム

:ラベル: `chap_recsys`

 **Shuai Zhang** ( *Amazon* )、 **Aston Zhang** ( *Amazon* )、および**Yi Tay** ( *Google* )

レコメンダー システムは業界で広く採用されており、私たちの日常生活のいたるところで使用されています。これらのシステムは、オンライン ショッピング サイト (例: amazon.com)、音楽/映画サービス サイト (例: Netflix や Spotify)、モバイル アプリケーション ストア (例: IOS アプリ ストアや Google Play)、オンラインなどの多くの分野で利用されています。ほんの数例を挙げると、広告。

レコメンダー システムの主な目的は、ユーザーが見るべき映画、読みたいテキスト、購入すべき製品などの関連アイテムを見つけて、楽しいユーザー エクスペリエンスを生み出すのを支援することです。さらに、レコメンダー システムは、オンライン小売業者が収益の増加を促進するために導入する最も強力な機械学習システムの 1 つです。レコメンダー システムは、プロアクティブな検索の労力を軽減し、検索したことのないオファーでユーザーを驚かせることにより、検索エンジンに代わるものです。多くの企業は、より効果的なレコメンダー システムの助けを借りて、競合他社よりも優位に立つことができました。そのため、レコメンダー システムは私たちの日常生活の中心であるだけでなく、一部の業界でも非常に不可欠なものとなっています。

この章では、レコメンダー システムの基礎と進歩について説明するとともに、利用可能なさまざまなデータ ソースを使用してレコメンダー システムを構築するための一般的な基本テクニックとその実装について説明します。具体的には、ユーザーが見込みのあるアイテムに与える可能性のある評価を予測する方法、アイテムの推奨リストを生成する方法、豊富な機能からクリックスルー率を予測する方法を学びます。これらのタスクは、実際のアプリケーションでは一般的なものです。この章を学習すると、古典的な手法だけでなく、より高度な深層学習ベースのモデルを使用して現実世界のレコメンデーション問題を解決する実践的な経験を得ることができます。
