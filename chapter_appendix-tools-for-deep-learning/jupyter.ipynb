
# Jupyter ノートブックの使用

:ラベル: `sec_jupyter`

このセクションでは、Jupyter Notebook を使用して、本書の各セクションのコードを編集および実行する方法について説明します。 :ref: `chap_installation`の説明に従って、Jupyter をインストールし、コードをダウンロードしたことを確認してください。 Jupyter についてさらに詳しく知りたい場合は、[ドキュメント](https://jupyter.readthedocs.io/en/latest/)内の優れたチュートリアルを参照してください。

## コードをローカルで編集して実行する

書籍のコードのローカル パスが`xx/yy/d2l-en/`であるとします。シェルを使用してディレクトリをこのパス ( `cd xx/yy/d2l-en` ) に変更し、コマンド`jupyter notebook`を実行します。ブラウザがこれを自動的に実行しない場合は、 http://localhost:8888 を開くと、Jupyter のインターフェイスと、 :numref: `fig_jupyter00`に示すように、書籍のコードを含むすべてのフォルダーが表示されます。

![](../img/jupyter00.png) :幅: `600px` :ラベル: `fig_jupyter00`

 Web ページに表示されているフォルダーをクリックすると、ノートブック ファイルにアクセスできます。通常、接尾辞「.ipynb」が付いています。簡潔にするために、一時的な「test.ipynb」ファイルを作成します。クリック後に表示される内容は :numref: `fig_jupyter01`に示されています。このノートブックには、マークダウン セルとコード セルが含まれています。マークダウン セルの内容には、「これはタイトルです」と「これはテキストです」が含まれます。コード セルには 2 行の Python コードが含まれています。 

![](../img/jupyter01.png) :幅: `600px` :ラベル: `fig_jupyter01`

マークダウン セルをダブルクリックして編集モードに入ります。新しいテキスト文字列「Hello world」を追加します。 :numref: `fig_jupyter02`に示すように、セルの最後にあります。

![](../img/jupyter02.png) :幅: `600px` :ラベル: `fig_jupyter02`

 :numref: `fig_jupyter03`に示すように、メニュー バーの "Cell" $\rightarrow$ "Run Cells" をクリックして、編集したセルを実行します。

![](../img/jupyter03.png) :幅: `600px` :ラベル: `fig_jupyter03`

実行後、マークダウン セルが :numref: `fig_jupyter04`に表示されます。

![](../img/jupyter04.png) :幅: `600px` :ラベル: `fig_jupyter04`

次に、コードセルをクリックします。 :numref: `fig_jupyter05`に示すように、コードの最後の行の後で要素を 2 で乗算します。

![](../img/jupyter05.png) :幅: `600px` :ラベル: `fig_jupyter05`

ショートカット (デフォルトでは「Ctrl + Enter」) を使用してセルを実行し、 :numref: `fig_jupyter06`から出力結果を取得することもできます。 

![](../img/jupyter06.png) :幅: `600px` :ラベル: `fig_jupyter06`

ノートブックにさらに多くのセルが含まれている場合は、メニュー バーの [カーネル] $\rightarrow$ [再起動してすべて実行] をクリックして、ノートブック全体のすべてのセルを実行できます。メニューバーの「ヘルプ」$\rightarrow$「キーボードショートカットの編集」をクリックすると、好みに応じてショートカットを編集できます。

## 高度なオプション

ローカル編集以外に、マークダウン形式でノートブックを編集することと、Jupyter をリモートで実行することという 2 つのことが非常に重要です。後者は、より高速なサーバーでコードを実行したい場合に重要です。 Jupyter のネイティブ ipynb 形式には、コンテンツとは無関係な多くの補助データ (主にコードの実行方法と場所に関連するデータ) が格納されるため、前者が重要になります。これは Git にとって混乱を招くため、コントリビューションのレビューが非常に困難になります。幸いなことに、マークダウン形式でのネイティブ編集という代替手段があります。
