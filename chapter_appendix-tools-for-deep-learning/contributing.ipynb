
# この本への寄稿

:label: `sec_how_to_contribute`

[読者](https://github.com/d2l-ai/d2l-en/graphs/contributors)による貢献は、この書籍の改善に役立ちます。タイプミス、古いリンク、引用を見逃したと思われる箇所、コードがエレガントに見えない箇所、または説明が不明瞭な箇所を見つけた場合は、ぜひ貢献して読者を支援してください。通常の本では、印刷間隔 (つまり、タイプミスの修正間隔) が数年かかることもありますが、この本に改善を反映するには通常、数時間から数日かかります。これはすべて、バージョン管理と継続的インテグレーション (CI) テストによって可能になります。これを行うには、GitHub リポジトリに[プル リクエスト](https://github.com/d2l-ai/d2l-en/pulls)を送信する必要があります。あなたのプル リクエストが作成者によってコード リポジトリにマージされると、あなたはコントリビュータになります。

## 軽微な変更の送信

最も一般的な貢献は、一文の編集またはタイプミスの修正です。 [GitHub リポジトリ](https://github.com/d2l-ai/d2l-en)でソース ファイルを見つけて、ファイルを直接編集することをお勧めします。たとえば、 [[ファイルの検索]](https://github.com/d2l-ai/d2l-en/find/master)ボタン (:numref: `fig_edit_file` ) を使用してファイルを検索し、ソース ファイル (マークダウン ファイル) を見つけることができます。次に、右上隅にある [このファイルを編集] ボタンをクリックして、マークダウン ファイルに変更を加えます。

![](../img/edit-file.png) :幅: `300px` :ラベル: `fig_edit_file`

完了したら、ページ下部の [ファイル変更の提案] パネルに変更の説明を入力し、[ファイル変更の提案] ボタンをクリックします。新しいページにリダイレクトされ、変更内容を確認できます (:numref: `fig_git_createpr` )。すべて問題がなければ、「Create pull request」ボタンをクリックしてプルリクエストを送信できます。

## 大きな変更の提案

テキストやコードの大部分を更新する予定がある場合は、この本で使用されている形式についてもう少し詳しく知る必要があります。ソース ファイルは[マークダウン形式](https://daringfireball.net/projects/markdown/syntax)に基づいており、数式、画像、章、引用の参照など、 [d2lbook](http://book.d2l.ai/user/markdown.html)パッケージによる一連の拡張子が付いています。任意のマークダウン エディターを使用してこれらのファイルを開いて変更を加えることができます。

コードを変更したい場合は、:numref: `sec_jupyter`で説明されているように、Jupyter Notebook を使用してこれらのマークダウン ファイルを開くことをお勧めします。これにより、変更を実行してテストできます。変更を送信する前に必ずすべての出力をクリアしてください。CI システムは更新したセクションを実行して出力を生成します。

一部のセクションでは、複数のフレームワーク実装をサポートする場合があります。新しいコード ブロックを追加する場合は、 `%%tab`を使用して先頭行にこのブロックをマークしてください。たとえば、PyTorch コード ブロックの場合は`%%tab pytorch` 、TensorFlow コード ブロックの場合は`%%tab tensorflow` 、すべての実装の共有コード ブロックの場合は`%%tab all` 。詳細については、 [`d2lbook`](http://book.d2l.ai/user/code_tabs.html)パッケージを参照してください。

## 大きな変更の送信

大きな変更を送信するには、標準の Git プロセスを使用することをお勧めします。簡単に言うと、プロセスは :numref: `fig_contribute`で説明されているように動作します。

![](http://d2l.ai/_images/contribute.svg) :label: `fig_contribute`

手順を詳しく説明します。すでに Git に精通している場合は、このセクションをスキップしてください。具体的にするために、投稿者のユーザー名は「astonzhang」であると仮定します。

###  Gitのインストール

Git オープン ソース ブックには、 [Git のインストール方法が](https://git-scm.com/book/en/v2)説明されています。これは通常、Ubuntu Linux で`apt install git`使用するか、macOS に Xcode 開発者ツールをインストールするか、GitHub の[デスクトップ クライアント](https://desktop.github.com)を使用して機能します。 GitHub アカウントをお持ちでない場合は、アカウントにサインアップする必要があります。

###  GitHub へのログイン

ブラウザに書籍のコード リポジトリの[アドレス](https://github.com/d2l-ai/d2l-en/)を入力します。 :numref: `fig_git_fork`の右上にある赤いボックスの`Fork` ] ボタンをクリックして、この本のリポジトリのコピーを作成します。これで*コピーが作成*され、好きなように変更できます。

![](../img/git-fork.png) :幅: `700px` :ラベル: `fig_git_fork`

これで、この本のコード リポジトリが、 :numref: `fig_git_forked`の左上に示される`astonzhang/d2l-en`などのユーザー名にフォーク (つまり、コピー) されます。

![](../img/git-forked.png) :幅: `700px` :ラベル: `fig_git_forked`

### リポジトリのクローン作成

リポジトリのクローンを作成するには (つまり、ローカル コピーを作成するには)、そのリポジトリ アドレスを取得する必要があります。 :numref: `fig_git_clone`の緑色のボタンがこれを表示します。このフォークを長期間維持する場合は、ローカル コピーがメイン リポジトリで最新であることを確認してください。現時点では、:ref: `chap_installation`の指示に従って開始してください。主な違いは、リポジトリの*独自のフォーク*をダウンロードしていることです。

![](../img/git-clone.png) :幅: `700px` :ラベル: `fig_git_clone`

### 編集とプッシュ

今度は本の編集です。 :numref: `sec_jupyter`指示に従って、Jupyter Notebook で編集するのが最善です。変更を加え、問題がないことを確認します。ファイル`~/d2l-en/chapter_appendix-tools-for-deep-learning/contributing.md`タイプミスを修正したと仮定します。その後、どのファイルを変更したかを確認できます。

この時点で、Git は`chapter_appendix-tools-for-deep-learning/contributing.md`ファイルが変更されたことを通知するプロンプトを表示します。

これが希望どおりであることを確認したら、次のコマンドを実行します。

変更されたコードは、リポジトリの個人用フォークに置かれます。変更の追加をリクエストするには、書籍の公式リポジトリに対してプル リクエストを作成する必要があります。

### プルリクエストの送信

:numref: `fig_git_newpr`に示すように、GitHub 上のリポジトリのフォークに移動し、[新しいプル リクエスト] を選択します。これにより、編集間の変更と書籍のメイン リポジトリの現在の内容を示す画面が開きます。

![](../img/git-newpr.png) :幅: `700px` :ラベル: `fig_git_newpr`

最後に、 :numref: `fig_git_createpr`に示すボタンをクリックしてプル リクエストを送信します。プル リクエストで行った変更を必ず説明してください。これにより、著者がそれをレビューしたり、本に統合したりすることが容易になります。変更内容に応じて、これはすぐに承認される場合もあれば、拒否される場合もあります。あるいは、変更について何らかのフィードバックが得られる可能性が高くなります。それらを組み込んだら、準備完了です。

![](../img/git-createpr.png) :幅: `700px` :ラベル: `fig_git_createpr`

## まとめ
- GitHub を使用してこの本に貢献できます。
- 軽微な変更については、GitHub 上のファイルを直接編集できます。
- 大きな変更を行う場合は、リポジトリをフォークしてローカルで編集し、準備ができてからのみ貢献してください。
- プル リクエストは、投稿がまとめられる方法です。巨大なプル リクエストを送信すると、理解して組み込むことが困難になるため、送信しないようにしてください。小さいものをいくつか送った方が良いでしょう。

## 演習
1. `d2l-ai/d2l-en`リポジトリにスターを付けてフォークします。
1. 改善が必要な点 (参照の欠落など) を見つけた場合は、プル リクエストを送信してください。
1. 通常は、新しいブランチを使用してプル リクエストを作成することをお勧めします。 [Git ブランチングを](https://git-scm.com/book/en/v2/Git-Branching-Branches-in-a-Nutshell)使用してそれを行う方法を学びましょう。
