
# 表記

:label: `chap_notation`

本書全体を通じて、次の表記規則に従います。これらのシンボルの一部はプレースホルダーですが、他のシンボルは特定のオブジェクトを参照していることに注意してください。一般的な経験則として、不定冠詞「a」は、多くの場合、そのシンボルがプレースホルダーであり、同様にフォーマットされたシンボルが同じタイプの他のオブジェクトを表すことができることを示します。たとえば、「$x$: スカラー」は小文字が一般にスカラー値を表すことを意味しますが、「$\mathbb{Z}$: 整数のセット」は特に記号 $\mathbb{Z}$ を指します。

## 数値オブジェクト
- $x$: スカラー
- $\mathbf{x}$: ベクトル
- $\mathbf{X}$: 行列
- $\mathsf{X}$: 一般テンソル
- $\mathbf{I}$: (特定の次元の)単位行列、つまり、すべての対角要素に $1$ があり、すべての非対角要素に $0$ がある正方行列
- $x_i$, $[\mathbf{x}]_i$: ベクトル $\mathbf{x}$ の $i^\mathrm{th}$ 要素
- $x_{ij}$, $x_{i,j}$,$[\mathbf{X}] *{ij}$, $[\mathbf{X}]* {i,j}$: 行列 $\ の要素mathbf{X}$ は行 $i$ と列 $j$ にあります。

## 集合論
- $\mathcal{X}$: セット
- $\mathbb{Z}$: 整数のセット
- $\mathbb{Z}^+$: 正の整数のセット
- $\mathbb{R}$: 実数のセット
- $\mathbb{R}^n$: 実数の $n$ 次元ベクトルの集合
- $\mathbb{R}^{a\times b}$: $a$ 行と $b$ 列を持つ実数の行列のセット
- $|\mathcal{X}|$: セット $\mathcal{X}$ のカーディナリティ (要素の数)
-  $\mathcal{A}\cup\mathcal{B}$: セット $\mathcal{A}$ と $\mathcal{B}$ の和集合
- $\mathcal{A}\cap\mathcal{B}$: 集合 $\mathcal{A}$ と $\mathcal{B}$ の積集合
- $\mathcal{A}\setminus\mathcal{B}$: $\mathcal{A}$ からの $\mathcal{B}$ の減算を設定します (属さない $\mathcal{A}$ の要素のみが含まれます) $\mathcal{B}$ に)

## 関数と演算子
- $f(\cdot)$: 関数
- $\log(\cdot)$: 自然対数 (底 $e$)
-  $\log_2(\cdot)$: $2$を底とする対数
- $\exp(\cdot)$: 指数関数
- $\mathbf{1}(\cdot)$: インジケータ関数。ブール引数が true の場合は $1$ に評価され、そうでない場合は $0$ に評価されます。
-  $\mathbf{1}_{\mathcal{X}}(z)$: 集合メンバーシップ指標関数。要素 $z$ が集合 $\mathcal{X}$ および $0$ に属する場合、$1$ と評価されます。さもないと
- $\mathbf{(\cdot)}^\top$: ベクトルまたは行列の転置
- $\mathbf{X}^{-1}$: 行列 $\mathbf{X}$ の逆行列
- $\odot$: アダマール (要素ごとの) 積
- $[\cdot, \cdot]$: 連結
- $|\cdot|_p$: $\ell_p$ ノルム
- $|\cdot|$: $\ell_2$ ノルム
- $\langle \mathbf{x}, \mathbf{y} \rangle$: ベクトル $\mathbf{x}$ と $\mathbf{y}$ の内積
- $\sum$: 要素のコレクションの合計
- $\prod$: 要素のコレクション上の積
- $\stackrel{\mathrm{def}}{=}$: 左側のシンボルの定義として主張される等価性

## 微積分
- $\frac{dy}{dx}$: $x$ に関する $y$ の導関数
- $\frac{\partial y}{\partial x}$: $x$ に関する $y$ の偏導関数
- $\nabla_{\mathbf{x}} y$: $\mathbf{x}$ に対する $y$ の勾配
- $\int_a^bf(x) ;dx$: $f$ の $x$ に対する $a$ から $b$ までの定積分
- $\int f(x) ;dx$: $x$ に関する $f$ の不定積分

## 確率と情報理論
- $X$: 確率変数
- $P$: 確率分布
- $X \sim P$: 確率変数 $X$ は分布 $P$ に従います
- $P(X=x)$: 確率変数 $X$ が値 $x$ を取るイベントに割り当てられる確率
- $P(X \mid Y)$: $Y$ が与えられた場合の $X$ の条件付き確率分布
- $p(\cdot)$: 分布 P に関連付けられた確率密度関数 (PDF)
-  ${E}[X]$: 確率変数 $X$ の期待値
- $X \perp Y$: 確率変数 $X$ と $Y$ は独立しています
- $X \perp Y \mid Z$: 確率変数 $X$ と $Y$ は、$Z$ が与えられると条件付きで独立します。
-  $\sigma_X$: 確率変数 $X$ の標準偏差
- $\mathrm{Var}(X)$: 確率変数 $X$ の分散、$\sigma^2_X$ に等しい
- $\mathrm{Cov}(X, Y)$: 確率変数 $X$ と $Y$ の共分散
- $\rho(X, Y)$: $X$ と $Y$ の間のピアソン相関係数、$\frac{\mathrm{Cov}(X, Y)}{\sigma_X \sigma_Y}$ と等しい
- $H(X)$: 確率変数 $X$ のエントロピー
- $D_{\mathrm{KL}}(P|Q)$: 分布 $Q$ から分布 $P$ までの KL 発散 (または相対エントロピー)
