
# 序文

ほんの数年前までは、大手企業や新興企業でインテリジェントな製品やサービスを開発しているディープラーニング科学者の集団は存在しませんでした。私たちがこの分野に入ったとき、機械学習は毎日の新聞の見出しを飾るほどではありませんでした。私たちの両親は、機械学習が何なのか、ましてや私たちがなぜ医学や法律の仕事よりも機械学習を好むのか全く知りませんでした。機械学習は青天の霹靂の学問分野であり、その産業上の重要性は、音声認識やコンピュータ ビジョンなどの限られた現実世界のアプリケーションに限定されていました。さらに、これらのアプリケーションの多くは非常に多くの専門知識を必要とするため、機械学習が 1 つの小さなコンポーネントである完全に別個の領域とみなされることがよくありました。当時、ニューラル ネットワーク (本書で焦点を当てている深層学習手法の前身) は、一般に時代遅れであると見なされていました。

ここ数年で、ディープ ラーニングは世界を驚かせ、コンピューター ビジョン、自然言語処理、自動音声認識、強化学習、生物医学情報学などのさまざまな分野で急速な進歩をもたらしました。さらに、実用的に興味深い非常に多くのタスクにおける深層学習の成功は、理論的な機械学習と統計の発展を促進することさえありました。これらの進歩により、私たちは今、これまでよりも高い自律性 (一部の企業が信じているほど自律性は低い) で自動運転する自動車、明確な質問をしてコードをデバッグする対話システム、そして世界を支配するソフトウェア エージェントを構築できるようになりました。かつては数十年先のことだと思われていた、囲碁などのボードゲームで最高の人間だ。すでに、これらのツールは産業や社会にこれまで以上に広範な影響を与えており、映画の制作方法や病気の診断方法を変え、天体物理学から生物学に至る基礎科学での役割も増大しています。

## この本について

この本は、深層学習を親しみやすくするという私たちの試みを表しており、*概念*、*コンテキスト*、*コードを*教えています。

### コード、数学、HTML を組み合わせた 1 つのメディア

コンピューティング テクノロジがその効果を最大限に発揮するには、それが十分に理解され、十分に文書化され、成熟し、よく保守されたツールによってサポートされている必要があります。重要なアイデアを明確に抽出し、新しい実務者に最新情報を提供するために必要なオンボーディング時間を最小限に抑える必要があります。成熟したライブラリは一般的なタスクを自動化する必要があり、サンプル コードにより、実践者がニーズに合わせて一般的なアプリケーションを簡単に変更、適用、拡張できるようにする必要があります。動的 Web アプリケーションを例に挙げます。 1990 年代には Amazon のような多くの企業がデータベース駆動型 Web アプリケーションの開発で成功を収めましたが、クリエイティブな起業家を支援するこのテクノロジーの可能性は、過去 10 年間ではるかに大きく実現されました。強力で十分に文書化されたフレームワーク。

単一のアプリケーションにはさまざまな分野が統合されているため、深層学習の可能性をテストすることには独特の課題が伴います。深層学習を適用するには、同時に理解する必要があります。(i) 問題を特定の方法で投げかける動機。 (ii) 特定のモデルの数学的形式。 (iii) モデルをデータに適合させるための最適化アルゴリズム。 (iv) いつモデルが目に見えないデータに一般化すると予想すべきかを示す統計原則と、モデルが実際に一般化されたことを証明する実際的な方法。 (v) モデルを効率的にトレーニングし、数値計算の落とし穴を回避し、利用可能なハードウェアを最大限に活用するために必要なエンジニアリング手法。問題を定式化するために必要な批判的思考スキル、問題を解決するための数学、およびそれらの解決策を実装するためのソフトウェア ツールの両方を 1 か所で教えることは、困難な課題となります。本書の目的は、実践者を目指す人に最新情報を提供するための統一されたリソースを提供することです。

私たちがこの書籍プロジェクトを開始したとき、同時に (i) 最新の状態を維持できるリソースはありませんでした。 (ii) 十分な技術的深みを備えた最新の機械学習実践の広範な内容をカバーしている。 (iii) 教科書に期待される品質と、実践的なチュートリアルに期待されるクリーンな実行可能なコードを交互に挟んだ解説。特定の深層学習フレームワークの使用方法 (例: TensorFlow の行列を使用した基本的な数値計算の実行方法)、または特定の手法の実装 (例: LeNet、AlexNet、ResNet などのコード スニペット) の散在するコード例が多数見つかりました。さまざまなブログ投稿や GitHub リポジトリにわたって。ただし、これらの例は通常、特定のアプローチを実装する*方法*に焦点を当てており、特定のアルゴリズムによる決定が行われる*理由*についての議論は省略されています。ウェブサイト[Distill](http://distill.pub)で公開されている魅力的なブログ投稿や個人のブログなど、特定のトピックに対処するために対話型リソースが散発的に出現することもありますが、これらは深層学習の選択されたトピックのみをカバーしており、関連するコードが欠けていることがよくあります。一方で、ディープラーニングの基礎に関する包括的な調査を提供する :citet: `Goodfellow.Bengio.Courville.2016`など、ディープラーニングの教科書がいくつか登場していますが、これらのリソースは説明と内容を結び付けていません。概念をコードで実現するため、読者はそれをどのように実装すればよいのか分からなくなることがあります。さらに、商用コースプロバイダーの料金の壁の後ろにあまりにも多くのリソースが隠されています。

私たちは、(i) 誰もが無料で利用できるリソースの作成に着手しました。 (ii) 実際に応用機械学習科学者になるための出発点を提供するのに十分な技術的な深みを提供する。 (iii) 実行可能なコードを含めて、実際に問題を解決する*方法を*読者に示します。 (iv) 私たちとコミュニティ全体の両方による迅速な更新を可能にします。 (v) 技術的な詳細について対話的に議論し、質問に答えるための[フォーラム](https://discuss.d2l.ai/c/5)によって補完されます。

これらの目標は矛盾することがよくありました。方程式、定理、引用は LaTeX で管理およびレイアウトするのが最適です。コードは Python で記述するのが最適です。また、Web ページは HTML と JavaScript でネイティブです。さらに、コンテンツには、実行可能コードとして、物理的な書籍として、ダウンロード可能な PDF として、そして Web サイトとしてインターネット上でアクセスできるようにしたいと考えています。これらの要求に適したワークフローはないようだったので、独自の (:numref: `sec_how_to_contribute` ) を組み立てることにしました。私たちはソースを共有し、コミュニティへの貢献を促進するために GitHub を使用することにしました。コード、方程式、テキストを混合するための Jupyter ノートブック。レンダリング エンジンとしての Sphinx。そしてディスカッションプラットフォームとしてのDiscourse。私たちのシステムは完璧ではありませんが、これらの選択は、競合する懸念事項の間で妥協点をもたらします。 *『Dive into Deep Learning』は、*そのような統合されたワークフローを使用して出版される最初の書籍になる可能性があると私たちは考えています。

### 実践から学ぶ

多くの教科書では、概念が次々に提示され、それぞれの概念が徹底的に詳細にカバーされています。たとえば、Chris Bishop の優れた教科書 :cite: `Bishop.2006` 、各トピックを徹底的に教えているため、線形回帰の章に到達するには、かなりの量の作業が必要です。専門家はまさにその徹底した内容でこの本を愛していますが、真の初心者にとっては、この特性により入門書としての有用性が制限されます。

この本では、ほとんどの概念を*ジャストインタイムで*教えます。言い換えれば、実際的な目的を達成するために必要なまさにその瞬間に概念を学ぶことになります。最初に線形代数や確率などの基本的な予備知識を教えるために少し時間を取りますが、より難解な概念について心配する前に、最初のモデルをトレーニングする満足感を味わってもらいたいと考えています。

基本的な数学的背景の短期集中コースを提供するいくつかの予備ノートブックとは別に、後続の各章では、適切な数の新しい概念を紹介し、実際のデータセットを使用したいくつかの自己完結型の実用的な例を提供します。これは組織的な課題となりました。一部のモデルは論理的に 1 つのノートブックにグループ化されている場合があります。また、いくつかのアイデアは、複数のモデルを連続して実行することで最も効果的に学習できる場合があります。一方で、 *1 つの実例と 1 つのノートブック*というポリシーに従うことには大きな利点があります。これにより、コードを活用して独自の研究プロジェクトをできるだけ簡単に開始できるようになります。ノートブックをコピーして、変更を開始するだけです。

全体を通して、必要に応じて、実行可能なコードと背景素材をインターリーブします。一般に、私たちはツールを十分に説明する前に、ツールを利用できるようにするという間違いを犯します (多くの場合、後で背景を説明します)。たとえば、*確率的勾配降下法が*なぜ役立つのかを説明したり、それが機能する理由を直感的に説明したりする前に、確率的勾配降下法を使用することがあります。これは、読者にキュレーションによる判断を信頼してもらうことを犠牲にして、専門家に問題を迅速に解決するために必要な弾薬を与えるのに役立ちます。

この本はディープラーニングの概念をゼロから教えます。場合によっては、最新の深層学習フレームワークでは通常ユーザーから隠蔽されるモデルの詳細を掘り下げることもあります。これは特に基本的なチュートリアルで取り上げられ、特定のレイヤーまたはオプティマイザーで起こるすべてのことを理解してもらいたいと考えています。このような場合、私たちは例の 2 つのバージョンを提示することがよくあります。1 つは、NumPy のような機能と自動微分のみに依存して、すべてを最初から実装する例で、もう 1 つは、NumPy のような高レベル API を使用して簡潔なコードを記述する、より実用的な例です。深層学習フレームワーク。一部のコンポーネントがどのように機能するかを説明した後、後続のチュートリアルでは高レベル API を使用します。

### 内容と構造

この本は、実際のシステムとアプリケーション (:numref: `fig_book_org` ) に焦点を当てた、予備知識、深層学習テクニック、および高度なトピックに焦点を当てた、およそ 3 つの部分に分かれています。

![](../img/book-org.svg) :label: `fig_book_org`
- **パート 1: 基本と準備**。 :numref: `chap_introduction`深層学習の入門を提供します。次に、 :numref: `chap_preliminaries`では、データの保存方法と操作方法、線形代数の基本概念に基づいたさまざまな数値演算の適用方法など、実践的な深層学習に必要な前提条件をすぐに理解できます。微積分と確率。 :numref: `chap_regression`と :numref: `chap_perceptrons` 、回帰や分類など、深層学習の最も基本的な概念と手法をカバーしています。線形モデル。多層パーセプトロン。そして過学習と正則化。
- **パート 2: 最新の深層学習テクニック**。 :numref: `chap_computation`深層学習システムの主要な計算コンポーネントを記述し、その後のより複雑なモデルの実装のための基礎を築きます。次に、:numref: `chap_cnn`および :numref: `chap_modern_cnn` 、最新のコンピューター ビジョン システムのバックボーンを形成する強力なツールである畳み込みニューラル ネットワーク (CNN) を紹介します。同様に、:numref: `chap_rnn`と :numref: `chap_modern_rnn` 、データ内の順次 (時間的) 構造を活用し、自然言語処理と時系列予測に一般的に使用されるモデルであるリカレント ニューラル ネットワーク (RNN) を導入します。 :numref: `chap_attention-and-transformers`では、ほとんどの自然言語処理タスクの主要なアーキテクチャとして RNN に取って代わる、いわゆる*アテンション メカニズム*に基づく比較的新しいクラスのモデルを導入します。これらのセクションでは、ディープ ラーニングの実践者によって広く使用されている最も強力で汎用的なツールについて説明します。
- **パート 3: スケーラビリティ、効率、およびアプリケーション**([オンラインで](https://d2l.ai)入手可能)。第 12 章では、深層学習モデルのトレーニングに使用されるいくつかの一般的な最適化アルゴリズムについて説明します。次に、第 13 章では、深層学習コードの計算パフォーマンスに影響を与えるいくつかの重要な要素を検討します。次に、第 14 章では、コンピューター ビジョンにおける深層学習の主な応用例を示します。最後に、第 15 章と第 16 章では、言語表現モデルを事前トレーニングし、それらを自然言語処理タスクに適用する方法を示します。

## まとめ

ディープラーニングはパターン認識に革命をもたらし、現在ではコンピューター ビジョン、自然言語処理、自動音声認識などのさまざまな分野の幅広いテクノロジーを強化するテクノロジーを導入しています。ディープラーニングをうまく適用するには、問題の設定方法、モデリングの基本的な数学、モデルをデータに適合させるためのアルゴリズム、およびそれらすべてを実装するためのエンジニアリング手法を理解する必要があります。この本は、散文、図、数学、コードをすべて 1 か所にまとめた包括的なリソースを提供します。この本に関連する質問をする (または回答する) には、フォーラム (https://discuss.d2l.ai) にアクセスしてください。すべてのノートブックは、 [D2L.ai Web サイト](https://d2l.ai)および[GitHub](https://github.com/d2l-ai/d2l-en)からダウンロードできます。
