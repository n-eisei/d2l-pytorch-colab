
# ビルダーズガイド

:label: `chap_computation`

巨大なデータセットと強力なハードウェアに加えて、優れたソフトウェア ツールもディープ ラーニングの急速な進歩に不可欠な役割を果たしてきました。 2007 年にリリースされた画期的な Theano ライブラリを皮切りに、柔軟なオープンソース ツールにより、研究者はモデルのプロトタイプを迅速に作成できるようになり、低レベルの変更を加える機能を維持しながら、標準コンポーネントをリサイクルする際の反復作業を回避できました。時間の経過とともに、ディープ ラーニングのライブラリは、ますます粗い抽象化を提供するように進化してきました。半導体設計者がトランジスタから論理回路までの指定からコードの作成に移行したのと同じように、ニューラル ネットワークの研究者は、個々の人工ニューロンの動作について考えることから、層全体の観点からネットワークを考えることに移行し、現在でははるかに粗い*ブロック*を念頭に置いてアーキテクチャを設計することが多くなりました。 。

これまで、いくつかの基本的な機械学習の概念を紹介し、完全に機能する深層学習モデルを構築してきました。前の章では、MLP の各コンポーネントを最初から実装し、高レベル API を活用して同じモデルを簡単に展開する方法も示しました。そこまで早く到達するために、ライブラリ*を呼び出しました*が、ライブラリ*がどのように機能するか*についてのより高度な詳細は省略しました。この章では、カーテンを剥がして、深層学習計算の主要なコンポーネント、つまりモデルの構築、パラメーターのアクセスと初期化、カスタム レイヤーとブロックの設計、ディスクへのモデルの読み取りと書き込み、劇的なパフォーマンスを達成するための GPU の活用を詳しく掘り下げます。スピードアップ。これらの洞察は、*エンド ユーザー*から*パワー ユーザー*に移行し、自分で発明したモデルを含め、より複雑なモデルを実装する柔軟性を維持しながら、成熟したディープ ラーニング ライブラリの利点を享受するために必要なツールを提供します。この章では新しいモデルやデータセットは紹介しませんが、これに続く高度なモデリングの章はこれらの手法に大きく依存しています。
