
# コンピュータビジョン

:ラベル: `chap_cv`

医療診断、自動運転車、カメラ監視、スマート フィルターなど、コンピューター ビジョン分野の多くのアプリケーションは、私たちの現在および将来の生活と密接に関係しています。近年、ディープラーニングは、コンピュータビジョンシステムのパフォーマンスを向上させる変革の力となっています。最先端のコンピューター ビジョン アプリケーションは、ディープ ラーニングとほぼ切り離せないものであると言えます。これを考慮して、この章ではコンピュータ ビジョンの分野に焦点を当て、最近学術界や産業界に影響を与えている手法とアプリケーションを調査します。

 :numref: `chap_cnn`と :numref: `chap_modern_cnn`では、コンピューター ビジョンで一般的に使用されるさまざまな畳み込みニューラル ネットワークを研究し、それらを単純な画像分類タスクに適用しました。この章の初めに、モデルの一般化を改善する可能性がある 2 つの方法、つまり*画像拡張*と*微調整について*説明し、それらを画像分類に適用します。ディープ ニューラル ネットワークは画像を複数のレベルで効果的に表現できるため、このようなレイヤーごとの表現は、*オブジェクト検出*、*セマンティック セグメンテーション*、*スタイル転送*などのさまざまなコンピューター ビジョン タスクでうまく使用されています。コンピューター ビジョンでレイヤーごとの表現を活用するという重要なアイデアに従って、オブジェクト検出のための主要なコンポーネントと技術から始めます。次に、画像のセマンティック セグメンテーションに*完全畳み込みネットワークを*使用する方法を示します。次に、スタイル転送テクニックを使用して、この本の表紙のような画像を生成する方法を説明します。最後に、この章とこれまでのいくつかの章の資料を 2 つの一般的なコンピューター ビジョン ベンチマーク データセットに適用して、この章を締めくくります。
